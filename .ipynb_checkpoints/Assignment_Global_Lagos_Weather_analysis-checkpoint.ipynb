```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
plt.style.use('ggplot')

%matplotlib inline
```

```python
glo = pd.read_csv('global_results.csv')
lag = pd.read_csv('lagos_results.csv')
lag
lag.describe()
```

```python
missing_values = round(lag['avg_temp'].mean(), 2)
lag['avg_temp'].fillna(missing_values, inplace = True)

missing_values2 = round(glo['avg_temp'].mean(), 2)
glo['avg_temp'].fillna(missing_values2, inplace = True)
lag.info()
glo.info()
```

```python
lag['SMA_5'] = lag.avg_temp.rolling(window=5).mean()
glo['SMA_5'] = glo.avg_temp.rolling(window=5).mean()
# lag.head()
glo.head()
```

```python
glo.describe()
```

## LINE CHART

```python
# year = np.array(list(lag['year']))
# avg_temp = np.array(list(lag['avg_temp']))
# avg_temp

fig, axs = plt.subplots(2,1, figsize=(30,15))
fig.suptitle("Plot1 weather report")

plot1 = axs[0]
plot2 = axs[1]

N = len(lag.index)

x=lag['avg_temp']
y=lag['year'] 

a=glo['avg_temp']
b=glo['year'] 

plot1.plot(y, x, color = "green", alpha = 0.7)
plot2.plot(b, a, color = "red", alpha = 0.7)
plt.show
plt.savefig('lagos_linechart.jpg')

# # ind = np.arange(N)
```

## SIMPLE MOVING AVERAGE

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig, axs = plt.subplots(2,1, figsize = (15,13))

plt.title(' Global and Lagos Plot 1')
# plt.xlabel('Year')
# plt.ylabel('Temperature')

plt1 =axs[0]
plt2 =axs[1]

plt.rcParams["figure.figsize"] = (15,13)

plt1.grid(True)
plt1.plot(lag['year'],lag['avg_temp'],label='Lagos weather')
plt1.plot(lag['year'],lag['SMA_5'],label='SMA 5 Months')
plt1.legend(loc=2)
plt.savefig('new_lagos_to_mv5.jpg')

# global

plt2.grid(True)
plt2.plot(glo['year'],glo['avg_temp'],label='Global weather report')
plt2.plot(glo['year'],glo['SMA_5'],label='SMA 5 Months')
plt2.legend(loc=2)
plt.savefig('new_glo_to_mv5.jpg')
```

```python
# global

plt2.grid(True)
plt2.plot(glo['avg_temp'],label='Global weather report')
plt2.plot(glo['SMA_5'],label='SMA 5 Months')
plt2.legend(loc=2)
# plt2.savefig('global_to_mv5.jpg')
```

## FILL BETWEEN FORM

```python
glo = pd.read_csv('global_results.csv')
lag = pd.read_csv('lagos_results.csv')


year = glo['year']
avg_temp = glo['avg_temp']

lag_year = lag['year']
lag_avg = lag['avg_temp']

# plt.plot(year,avg_temp, color = 'green', linestyle = '--', label = 'Global Results')
# plt.plot(year, lag_avg, color = 'red', linestyle = '--', label = 'Lagos Results')

overall_median_lag = 26.66
overall_median_glo = 8.72

# plt.fill_between(year, avg_temp, overall_median_glo, alpha = 0.3)
# plt.fill_between(year, lag_avg, overall_median_lag, alpha = 0.3)

plt.rcParams["figure.figsize"] = (15,7)
plt.fill_between(glo['year'], glo['avg_temp'], alpha = 0.5, color = 'red', label = 'Global Moving Report')
plt.fill_between(lag['year'], lag['avg_temp'], alpha = 0.5, color = 'green', label = 'Lagos Moving Report')
plt.legend()


plt.title('Fill-Between for Lagos and Global')
plt.xlabel('Years')
plt.ylabel('Moving Average')
plt.tight_layout

plt.legend(loc=2)
plt.savefig('Fill between2.jpg')

plt.show()
```

## FILL BETWEEN FORM WITH SET MEDIAN

```python

fig,axs = plt.subplots(2,1, figsize = (15,10))

plt1 = axs[0]
plt2 = axs[1]

glo = pd.read_csv('global_results.csv')
lag = pd.read_csv('lagos_results.csv')


year = glo['year']
avg_temp = glo['avg_temp']

lag_year = lag['year']
lag_avg = lag['avg_temp']

# plt1=plt
# plt.plot(year,avg_temp, color = 'green', linestyle = '--', label = 'Global Results')
# plt.plot(year, lag_avg, color = 'red', linestyle = '--', label = 'Lagos Results')

overall_median_lag = 26.66
overall_median_glo = 8.72

plt2.set_title('Fill-Between for Lagos with set median of 26.66')
plt1.set_title('Fill-Between for Global Results with set median of 8.72')
plt.xlabel('Years')
plt.ylabel('Moving Average')
plt.tight_layout



# plt.fill_between(year, avg_temp, overall_median_glo, alpha = 0.3)
# plt.fill_between(year, lag_avg, overall_median_lag, alpha = 0.3)

plt.rcParams["figure.figsize"] = (15,7)
plt1.fill_between(glo['year'], glo['avg_temp'],overall_median_glo, linestyle = '-', alpha = 0.25, color = 'red', label = 'Global Moving Report')
plt2.fill_between(lag['year'], lag['avg_temp'], overall_median_lag, linestyle = '--', alpha = 0.5, color = 'green', label = 'Lagos Moving Report')
plt.legend(loc =1)


# plt.title('Fill-Between for Lagos with set median of 26.66')
# plt.xlabel('Years')
# plt.ylabel('Moving Average')
# plt.tight_layout

plt2.legend(loc=2)
plt.savefig('Lagos_fill_between_median26.jpg')

plt.show()
```
