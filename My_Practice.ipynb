```python
x = (1,2,3,4,5,6,7)
y= [4,2,4,2,5,6,8]
z = [random.randint(8, 50) for _ in range (7)]

plt.plot(x,y)
plt.show()
plt.style.use('ggplot')
plt.scatter(x,y)
# plt.show()

plt.xlabel('Frequency')
plt.ylabel('Time')
plt.title('Lessons Freestyle')

plt.plot(x,y, color = 'green', alpha = 1, linewidth = 3, label = 'Boys')
plt.plot(x,z, color ='blue', alpha = 0.3, linewidth = 1, label = 'girls')
plt.legend(loc = 'upper right')
plt.show()
```
