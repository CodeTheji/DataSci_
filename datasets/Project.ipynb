## The moving Average

```python
data_nig['SMA_5'] = data_nig.unemployment.rolling(window=5).mean()
data_gh['SMA_5'] = data_gh.unemployment.rolling(window=5).mean()
data_nig.head(20)
```

## LINE CHART

```python
fig, axs = plt.subplots(2,1, figsize=(20,10))
fig.suptitle("PLOT OF THE UNEMPLOYMENT RATE")

plot1 = axs[0]
plot2 = axs[1]

N = len(data_nig.index)

plot1.set_title('Nigeria')
plot2.set_title('Ghana')

# plot1.tight_layout
# plot2.tight_layout

plot1.set_xlabel('year')
plot1.set_ylabel('unemployment')

plot2.set_xlabel('year')
plot2.set_ylabel('unemployment')

x=data_nig['unemployment']
y=data_nig['year'] 

a=data_gh['unemployment']
b=data_gh['year'] 



# plot1.plot(y, x, color = "green", alpha = 0.7, label = 'Nigeria unemployment rate')
# plot2.plot(b, a, color = "red", alpha = 0.7, label = 'Ghana unemployment rate')
# plt.show

plot1.plot(data_nig.year, data_nig.unemployment, color = 'red' , alpha = 0.7, linewidth = 2, label = 'Nigeria unemployment rate')
plot1.set_xticks(np.arange(min(data_nig.year), max(data_nig.year)+1, 1.0))

plot2.plot(data_gh.year, data_gh.unemployment, color = 'blue' , alpha = 0.7, linewidth = 2, label = 'Ghana unemployment rate')
plot2.set_xticks(np.arange(min(data_gh.year), max(data_gh.year)+1, 1.0))

plot1.legend()
# plot1.legend(loc =2)
plot2.legend(loc =2)

plt.show()

# plt.savefig('lagos_linechart.jpg')
```
