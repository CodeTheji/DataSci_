```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.DataFrame({"x":x, "y":y})

sns.lmplot("x","y", data = data, fit_reg=True)
```

```python
plt.xlabel('Night')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,color='green',alpha = 1, linewidth=4, label = 'Osun')
plt.plot(x,z,color='blue',alpha = 0.2, linewidth=2, label = 'Lagos')
plt.legend(loc = "upper right")
plt.show()
```

## TO COMBINE THE WHOLE GRAPH

```python
#To plot for french first rename the fig, axs = plt.subplots(1,2, the two to 3 figsize=(10,2)) -- i changed the figsize to (15,5)
fig, axs = plt.subplots(1,4, figsize=(15,5))
fig.suptitle("Test scores for students")
plot1 = axs[0]
plot2 = axs[1]
plot3 = axs[2]
plot4 = axs[3]

plot1.set_title("Math Scores")
plot2.set_title("English Scores")
plot3.set_title("French Scores")
plot4.set_title("Combined Scores")
#fig.title("")

study_hrs = [1,2,3,4,5,6,7]
math = [50,51,52,48,47,49,46]
english = [43,42,40,44,33,35,37]
french = [45,48,48,46,40,42,41]

plot1.plot(study_hrs, math, color = "green", alpha = 0.7)
plot2.plot(study_hrs, english, color = "cyan", alpha = 0.7)
plot3.plot(study_hrs, french, color = "red", alpha = 0.7)

plot4.plot(study_hrs, math, color = "green", alpha = 0.7, label = "maths")
plot4.plot(study_hrs, english, color = "cyan", alpha = 0.7, label = "english")
plot4.plot(study_hrs, french, color = "red", alpha = 0.7, label = "french")

plot4.legend() # to show the keys of the line , Note you must put "label ="
plt.show
```

```python
## To convert the bar to horizontal
N = 5 
menMeans = np.array((20,35,30,35,27))
womenMeans = np.array((25,32,34,20,25))
womenMeans = np.array((15,39,30,20,19))

menStd = (2,3,4,1,2)
womenStd = (3,5,2,3,3)

ind = np.arange(N)  # the x Locations for the groups
width = 0.3 # the width of the bars: can also len(x) sequence
men = ["Ade", "Shayo", "Bolu", "Gide", "Jemile"]
subject = ["Math", "Science", "History", "Chemistry", "Biology"]

plt.bar(ind - 0.3, menMeans, color = ['y', 'r', 'g'], width = width, label = 'Mean Score')
plt.bar(ind + 0.1, menStd, color = ['cyan'], width = width, label = 'Score STD')
plt.plot(ind + 0.1, menStd, label = 'Score STD')

plt.xticks(ind - 0.1, men) # Xticks is ussed to replace values in the x-axis

plt.xlabel('student name')
plt.ylabel('Score Values')
plt.legend()
plt.show()
```
