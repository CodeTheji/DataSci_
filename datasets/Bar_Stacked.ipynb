```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 5
menMeans = np.array((20,35,30,35,27))
womenMeans = np.array((25,32,34,20,25))
childrenMeans = np.array((10,20,15,65,9))
menStd = (2,3,4,1,2)
womenStd = (3,5,2,3,3)
ind = np.arange(N) #the x locations for the groups
width = 0.35 # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, menMeans, width, yerr=menStd)
p2 = plt.bar(ind, womenMeans, width, bottom = menMeans, yerr=womenStd)

p3 = plt.bar(ind, childrenMeans, width, bottom = menMeans, yerr=womenMeans)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ("G1", "G2", "G3", "G4", "G5"))
# plt.yticks(np.arange(0,81,10))
plt.legend((p1[0], p2[0]), ('Men', 'Women', 'Children'))
plt.show()
```
