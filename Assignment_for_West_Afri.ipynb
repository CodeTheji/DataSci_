```python
pd.set_option('max_column', 10)
df = pd.read_csv('west_african_countries.csv')
plt.rcParams["figure.figsize"] = (15,8) # fixing the width of the graph
df

plt.scatter(x=df['LAT1'], 
y=df['LON1'], 
s=df['Population1']/10000, 
alpha=0.5, 
c=df['Color'])

chart title, axis labels
plt.title('LAT, LON, Population Bubble')
plt.xlabel('Latitude')
plt.ylabel('Longitude')

x,y = df['LAT1'], df['LON1']

for i, txt in enumerate(df['Country']):
    plt.annotate(txt, (x[i], y[i]))
    print(i, txt, x[i], y[i], df['Population1'][i], df['Color'][i])
    
# Legend
country_list = list(df['Country'])
bubble_color_list = list(df['Color'])
l = []

for i in range(0, len(df.index)):
    l.append(mpatches.Patch(color=bubble_color_list[i], alpha=0.5, label=country_list[i]))
    
plt.legend(handles=l, loc=(1.05,0))
```
