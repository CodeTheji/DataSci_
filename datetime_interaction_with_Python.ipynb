```python
import requests
import datetime
from dateutil.parser import parse
```

```python
days = 10
response = requests.get(f'http://checklight.pythonanywhere.com/get_readings/1x0d001/{days}/')
```

```python
print(response.json())
```

```python
#### GET RAW DATA IN JSON FORMAT

raw_data = response.json()
```

```python
start_date = parse(input("Enter start date in format(YYYY-MM-DD) : "))
end_date = parse(input("Enter end date in format(YYYY-MM-DD) : "))

for data in raw_data.get("streets"):
    #data = parse(data.get("time"))
    date = datetime.datetime.strptime(data.get("time"), "%b. %d, %Y, %I:%M %p.")
    if start_date <= date and end_date >= date:
        print(date)
```

```python
found = []
total_hours = 0

for data in raw_data.get("streets"):
    date = datetime.datetime.strptime(data.get("time"), "%b. %d, %Y, %I:%M %p.")
    
    if data.get("status") == 1 and len(found) <1:
        
        found.append(date)
        print(date, data.get("status"))
        
    elif data.get("status") == 0 and len(found) == 1:
        found.append(date)
        print(date, data.get("status"))
        
    elif len(found) == 2:
        time_between = (found[1] - found[0]).total_seconds()/3600
        total_hours += time_between
        
        print(time_between, "\n")
        found.clear()
        
print("Total hours of light = ", total_hours)
```
